__[문제 4: 여러 손님 주문 처리]__

여러 손님들의 주문을 처리할 수 있도록 시스템을 확장하세요.

요구사항:
- 손님이 "종료"를 입력할 때까지 주문을 계속 받습니다.
- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
# stock = 현재 재고
# sales = 판매 현황
stock = {"팥붕어빵": 10 , "슈크림붕어빵": 8, "초코붕어빵": 5}
sales = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}

# 무한반복 가능
while True:
    print("🥐 어떤 붕어빵을 주문하시겠습니까")
    print("팥붕어빵 | 슈크림붕어빵 | 초코붕어빵 | 종료")
    fish_order = input()

    # 주문한 붕어빵 종류가 지정해 놓은 붕어빵 중에 있을 시
    if fish_order == "팥붕어빵" or fish_order == "슈크림붕어빵" or fish_order == "초코붕어빵":
        print("몇 개를 주문하시겠습니까")
        fish_count = int(input())

        print("🐟 주문 내역:", fish_order, fish_count, "개")

        # 판매 가능
        # 주문한 붕어빵의 갯수가 현재 붕어빵 재고보다 많을 시
        if fish_order in stock and stock[fish_order] >= fish_count:
            print("🫶 주문하신 붕어빵이 준비되었습니다")
            stock[fish_order] = stock[fish_order] - fish_count # 현재 재고 -= 판매 갯수
            sales[fish_order] = sales[fish_order] + fish_count # 판매 현황 += 판매 갯수
            print("🫶 판매완료")
            print("현재 재고:", stock)
            print("현재 판매량:", sales)
            print("-" * 30)

        # 판매 불가
        # 주문한 붕어빵의 갯수가 현재 붕어빵 재고보다 많을 시
        else:
            print("🥲 재고가 부족합니다")
            print("-" * 30)

    # 종료 선택시 주문 종료됨
    elif fish_order == "종료":
        print("🥐 방문해주셔서 감사합니다 🥐")
        break

    # 주문한 붕어빵 종류가 지정해 놓은 붕어빵 중에 없을 시 / 잘못된 붕어빵을 입력했을 시
    else:
        print("🥲 메뉴에 없는 붕어빵입니다")
        print("-" * 30)

```

__[문제 5: 관리자 모드]__

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 관리자 모드
# fish_inventory = 추가할 붕어빵 이름
# fish_plus = 추가할 붕어빵 갯수

while True:
    print("😈 관리자모드 입니다 😈")
    print("추가할 붕어빵 종류를 말씀해주세요")
    print("팥붕어빵 | 슈크림붕어빵 | 초코붕어빵 | 종료")
    fish_inventory = input()

    if fish_inventory == "종료":
        print("관리자 모드를 종료합니다")
        break

    # 메뉴에 있으면 추가
    if fish_inventory in stock:
        fish_plus = int(input("몇 개를 추가하시겠습니까: "))
        # 현재 재고 딕셔너리에 입력받은 값 만큼 붕어빵을 추가함
        stock[fish_inventory] += fish_plus
        print("현재 재고:", stock)
        print("-" * 30)

    # 메뉴에 없는 붕어빵 입력 시 / 잘못된 문자 입력 시
    else:
        print("🥲 메뉴에 없는 붕어빵입니다")
        print("-" * 30)
```

__[문제 6: 매출 확인]__

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# price = 가격 딕셔너리
price = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}

print("💸 가격")
for bread, p in price.items():
    print(f" - {bread}: {p}원")

# 판매량 = 붕어빵 이름 : 갯수
print("📊 현재 판매량")
for bread, count in sales.items():
    print(f" - {bread}: {count}개")

# 총매출 = 가격 * 판매량
total = (
    price["팥붕어빵"] * sales["팥붕어빵"]
    + price["슈크림붕어빵"] * sales["슈크림붕어빵"]
    + price["초코붕어빵"] * sales["초코붕어빵"]
)

print("💰 총매출:", total, "원")
```
