```python
# stock = 현재 재고
# sales = 판매 현황
stock = {"팥붕어빵": 10 , "슈크림붕어빵": 8, "초코붕어빵": 5}
sales = {"팥붕어빵": 0, "슈크림붕어빵": 0, "초코붕어빵": 0}

# 무한반복 가능
# "관리자" 입력 시 관리자 페이지로 진입
# "판매현황" 입력 시 판매현황 확인 가능
while True:
    print("🥐 어떤 붕어빵을 주문하시겠습니까")
    print("🫘팥붕어빵 | 🧁슈크림붕어빵 | 🍫초코붕어빵 | ❌종료")
    fish_order = input()

    # 주문한 붕어빵 종류가 지정해 놓은 붕어빵 중에 있을 시
    if fish_order == "팥붕어빵" or fish_order == "슈크림붕어빵" or fish_order == "초코붕어빵":
        print("🥨 몇 개를 주문하시겠습니까")
        fish_count = int(input())

        print("🐟 주문 내역:", fish_order, fish_count, "개")

        # 판매 가능
        # 주문한 붕어빵의 갯수가 현재 붕어빵 재고보다 많을 시
        if fish_order in stock and stock[fish_order] >= fish_count:
            print("🫶 주문하신 붕어빵이 준비되었습니다")
            stock[fish_order] = stock[fish_order] - fish_count # 현재 재고 -= 판매 갯수
            sales[fish_order] = sales[fish_order] + fish_count # 판매 현황 += 판매 갯수
            print("🫶 판매완료")
            # print("현재 재고:", stock)
            # print("현재 판매량:", sales)
            print("-" * 30)

        # 판매 불가
        # 주문한 붕어빵의 갯수가 현재 붕어빵 재고보다 많을 시
        else:
            print("🥲 재고가 부족합니다")
            print("-" * 30)

    # 종료 선택시 주문 종료됨
    elif fish_order == "종료":
        print("🥐 방문해주셔서 감사합니다 🥐")
        break

    # 관리자페이지
    elif fish_order == "관리자":
        # fish_inventory = 추가할 붕어빵 이름
        # fish_plus = 추가할 붕어빵 갯수

        while True:
            print("😈 관리자모드 입니다 😈")
            print("추가할 붕어빵 종류를 말씀해주세요")
            print("팥붕어빵 | 슈크림붕어빵 | 초코붕어빵 | 종료")
            fish_inventory = input()

            if fish_inventory == "종료":
                print("관리자 모드를 종료합니다")
                print("-" * 30)
                break

            # 메뉴에 있으면 추가
            if fish_inventory in stock:
                fish_plus = int(input("몇 개를 추가하시겠습니까: "))
                # 현재 재고 딕셔너리에 입력받은 값 만큼 붕어빵을 추가함
                stock[fish_inventory] += fish_plus
                print("현재 재고:", stock)
                print("-" * 30)

            # 메뉴에 없는 붕어빵 입력 시 / 잘못된 문자 입력 시
            else:
                print("🥲 메뉴에 없는 붕어빵입니다")
                print("-" * 30)

    elif fish_order == "판매현황":
        # price = 가격 딕셔너리
        price = {"팥붕어빵": 1000, "슈크림붕어빵": 1200, "초코붕어빵": 1500}

        # 판매량 = 붕어빵 이름 : 갯수
        print("💸 가격")
        for bread, p in price.items():
            print(f" - {bread}: {p}원")

        # 판매량 = 붕어빵 이름 : 갯수
        print("📊 현재 판매량")
        for bread, count in sales.items():
            print(f" - {bread}: {count}개")

        # 총매출 = 가격 * 판매량
        total = sum(price[bread] * sales[bread] for bread in sales)
        print("💰 총매출:", total, "원")
        print("🍞" * 30)



    # 주문한 붕어빵 종류가 지정해 놓은 붕어빵 중에 없을 시 / 잘못된 붕어빵을 입력했을 시
    else:
        print("🥲 메뉴에 없는 붕어빵입니다")
        print("-" * 30)

```
