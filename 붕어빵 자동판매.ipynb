```python
# stock = ν„μ¬ μ¬κ³ 
# sales = νλ§¤ ν„ν™©
stock = {"ν¥λ¶•μ–΄λΉµ": 10 , "μν¬λ¦Όλ¶•μ–΄λΉµ": 8, "μ΄μ½”λ¶•μ–΄λΉµ": 5}
sales = {"ν¥λ¶•μ–΄λΉµ": 0, "μν¬λ¦Όλ¶•μ–΄λΉµ": 0, "μ΄μ½”λ¶•μ–΄λΉµ": 0}

# λ¬΄ν•λ°λ³µ κ°€λ¥
# "κ΄€λ¦¬μ" μ…λ ¥ μ‹ κ΄€λ¦¬μ νμ΄μ§€λ΅ μ§„μ…
# "νλ§¤ν„ν™©" μ…λ ¥ μ‹ νλ§¤ν„ν™© ν™•μΈ κ°€λ¥
while True:
    print("π¥ μ–΄λ–¤ λ¶•μ–΄λΉµμ„ μ£Όλ¬Έν•μ‹κ² μµλ‹κΉ")
    print("π«ν¥λ¶•μ–΄λΉµ | π§μν¬λ¦Όλ¶•μ–΄λΉµ | π«μ΄μ½”λ¶•μ–΄λΉµ | βμΆ…λ£")
    fish_order = input()

    # μ£Όλ¬Έν• λ¶•μ–΄λΉµ μΆ…λ¥κ°€ μ§€μ •ν•΄ λ†“μ€ λ¶•μ–΄λΉµ μ¤‘μ— μμ„ μ‹
    if fish_order == "ν¥λ¶•μ–΄λΉµ" or fish_order == "μν¬λ¦Όλ¶•μ–΄λΉµ" or fish_order == "μ΄μ½”λ¶•μ–΄λΉµ":
        print("π¥¨ λ‡ κ°λ¥Ό μ£Όλ¬Έν•μ‹κ² μµλ‹κΉ")
        fish_count = int(input())

        print("π μ£Όλ¬Έ λ‚΄μ—­:", fish_order, fish_count, "κ°")

        # νλ§¤ κ°€λ¥
        # μ£Όλ¬Έν• λ¶•μ–΄λΉµμ κ°―μκ°€ ν„μ¬ λ¶•μ–΄λΉµ μ¬κ³ λ³΄λ‹¤ λ§μ„ μ‹
        if fish_order in stock and stock[fish_order] >= fish_count:
            print("π«¶ μ£Όλ¬Έν•μ‹  λ¶•μ–΄λΉµμ΄ μ¤€λΉ„λμ—μµλ‹λ‹¤")
            stock[fish_order] = stock[fish_order] - fish_count # ν„μ¬ μ¬κ³  -= νλ§¤ κ°―μ
            sales[fish_order] = sales[fish_order] + fish_count # νλ§¤ ν„ν™© += νλ§¤ κ°―μ
            print("π«¶ νλ§¤μ™„λ£")
            # print("ν„μ¬ μ¬κ³ :", stock)
            # print("ν„μ¬ νλ§¤λ‰:", sales)
            print("-" * 30)

        # νλ§¤ λ¶κ°€
        # μ£Όλ¬Έν• λ¶•μ–΄λΉµμ κ°―μκ°€ ν„μ¬ λ¶•μ–΄λΉµ μ¬κ³ λ³΄λ‹¤ λ§μ„ μ‹
        else:
            print("π¥² μ¬κ³ κ°€ λ¶€μ΅±ν•©λ‹λ‹¤")
            print("-" * 30)

    # μΆ…λ£ μ„ νƒμ‹ μ£Όλ¬Έ μΆ…λ£λ¨
    elif fish_order == "μΆ…λ£":
        print("π¥ λ°©λ¬Έν•΄μ£Όμ…”μ„ κ°μ‚¬ν•©λ‹λ‹¤ π¥")
        break

    # κ΄€λ¦¬μνμ΄μ§€
    elif fish_order == "κ΄€λ¦¬μ":
        # fish_inventory = μ¶”κ°€ν•  λ¶•μ–΄λΉµ μ΄λ¦„
        # fish_plus = μ¶”κ°€ν•  λ¶•μ–΄λΉµ κ°―μ

        while True:
            print("π κ΄€λ¦¬μλ¨λ“ μ…λ‹λ‹¤ π")
            print("μ¶”κ°€ν•  λ¶•μ–΄λΉµ μΆ…λ¥λ¥Ό λ§μ”€ν•΄μ£Όμ„Έμ”")
            print("ν¥λ¶•μ–΄λΉµ | μν¬λ¦Όλ¶•μ–΄λΉµ | μ΄μ½”λ¶•μ–΄λΉµ | μΆ…λ£")
            fish_inventory = input()

            if fish_inventory == "μΆ…λ£":
                print("κ΄€λ¦¬μ λ¨λ“λ¥Ό μΆ…λ£ν•©λ‹λ‹¤")
                print("-" * 30)
                break

            # λ©”λ‰΄μ— μμΌλ©΄ μ¶”κ°€
            if fish_inventory in stock:
                fish_plus = int(input("λ‡ κ°λ¥Ό μ¶”κ°€ν•μ‹κ² μµλ‹κΉ: "))
                # ν„μ¬ μ¬κ³  λ”•μ…”λ„λ¦¬μ— μ…λ ¥λ°›μ€ κ°’ λ§νΌ λ¶•μ–΄λΉµμ„ μ¶”κ°€ν•¨
                stock[fish_inventory] += fish_plus
                print("ν„μ¬ μ¬κ³ :", stock)
                print("-" * 30)

            # λ©”λ‰΄μ— μ—†λ” λ¶•μ–΄λΉµ μ…λ ¥ μ‹ / μλ»λ λ¬Έμ μ…λ ¥ μ‹
            else:
                print("π¥² λ©”λ‰΄μ— μ—†λ” λ¶•μ–΄λΉµμ…λ‹λ‹¤")
                print("-" * 30)

    elif fish_order == "νλ§¤ν„ν™©":
        # price = κ°€κ²© λ”•μ…”λ„λ¦¬
        price = {"ν¥λ¶•μ–΄λΉµ": 1000, "μν¬λ¦Όλ¶•μ–΄λΉµ": 1200, "μ΄μ½”λ¶•μ–΄λΉµ": 1500}

        # νλ§¤λ‰ = λ¶•μ–΄λΉµ μ΄λ¦„ : κ°―μ
        print("π’Έ κ°€κ²©")
        for bread, p in price.items():
            print(f" - {bread}: {p}μ›")

        # νλ§¤λ‰ = λ¶•μ–΄λΉµ μ΄λ¦„ : κ°―μ
        print("π“ ν„μ¬ νλ§¤λ‰")
        for bread, count in sales.items():
            print(f" - {bread}: {count}κ°")

        # μ΄λ§¤μ¶ = κ°€κ²© * νλ§¤λ‰
        total = sum(price[bread] * sales[bread] for bread in sales)
        print("π’° μ΄λ§¤μ¶:", total, "μ›")
        print("π" * 30)



    # μ£Όλ¬Έν• λ¶•μ–΄λΉµ μΆ…λ¥κ°€ μ§€μ •ν•΄ λ†“μ€ λ¶•μ–΄λΉµ μ¤‘μ— μ—†μ„ μ‹ / μλ»λ λ¶•μ–΄λΉµμ„ μ…λ ¥ν–μ„ μ‹
    else:
        print("π¥² λ©”λ‰΄μ— μ—†λ” λ¶•μ–΄λΉµμ…λ‹λ‹¤")
        print("-" * 30)

```
